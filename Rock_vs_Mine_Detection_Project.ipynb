Importing the dependencies to projects

```python
# numpy use to handel arrys in project
import numpy as np

# pandas use uplod data set and manage data set as table ( Data Frame )
import pandas as pd

# need to splite my data set in to Train / Test
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

# use for messure accuracy of my model
from sklearn.metrics import accuracy_score
```

Data Collection and Data Processing Part

```python
# import sonar data in to pandas dataFarame
sonar_dataset = pd.read_csv('/content/sonar data.csv')
```

```python
# check data set first 5
sonar_dataset.head()
```

```python
# check data set last 5
sonar_dataset.tail()
```

```python
# see number of rows and coloum and other data
sonar_dataset.shape
```

```python
# discribe data set
# 25% -  25% of data less that this value
sonar_dataset.describe()
```

```python
# data set data types
# mine type data - 111
# rock type data 96
# iloc → means “integer-location based indexing” in pandas.
#        It lets you select rows and columns by position (number index) instead of by label.
# : → means “all rows”.
# -1 → means “the last column” (because Python supports negative indexing).

sonar_dataset.iloc[:, -1].value_counts()
```

```python
# grop by data set accoding to last coloum and get mean value of it
sonar_dataset.groupby(sonar_dataset.iloc[:, -1]).mean(numeric_only=True)
```

```python
# sapperate data and lables
# x - data set
# y - finel value data set
x = sonar_dataset.drop(sonar_dataset.columns[-1], axis=1)
y = sonar_dataset[sonar_dataset.columns[-1]]
```

```python
print(x)
```

```python
print(y)
```

Splitting test and trin data


```python
# split data in to train and test data to tarin model and test it
# x train -  part of x data set to use as train
# y train -  part of y data set to use as train ( final result values )
# 10% use as test dataset
x_train ,x_test , y_train , y_test = train_test_split(x,y,test_size=0.1, stratify=y , random_state=1)
```

```python
# check shape of test and traing data sets
print(x.shape, x_train.shape, x_test.shape)
```

Model traing using Logistic Regression

```python
model = LogisticRegression()
```

```python
# traing model using train data set
model.fit(x_train,y_train)
```

```python
# Model Evaluation
training_data_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(training_data_prediction,y_train)
print('Accuracy on training data : ', training_data_accuracy)

test_data_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(test_data_prediction,y_test)
print('Accuracy on training data : ', test_data_accuracy)
```
